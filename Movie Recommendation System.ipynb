```python
import numpy as np
import pandas as pd
```

```python
column_names=['user_id','item_id','ratings','timestamp']
```

```python
#Reading user ratings data and appending column names to it
data=pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
data.head(5)
```

```python
movie_names=pd.read_csv("Movie_Id_Titles")
movie_names.head()
```

```python
movie=pd.merge(data,movie_names,on='item_id')
```

```python
movie_mat=movie.pivot_table(index='user_id',columns='title').ratings
```

```python
movie_mat.head()
```

```python
ratings=pd.DataFrame(movie.groupby('title')['ratings'].mean().sort_values(ascending=False))
```

```python
ratings.head()
```

```python
ratings['number_of_ratings']=pd.DataFrame(movie.groupby('title')['ratings'].count())
```

```python
ratings.head()
```

```python
movie_name=input("Enter the movie name");
```

```python
print("You entered",movie_name)
```

```python
movie_ratings=pd.DataFrame(movie_mat[movie_name])
```

```python
movie_ratings.head()
```

```python
similar_to_movie = movie_mat.corr(method='pearson')
```

```python
corr_movie=pd.DataFrame(similar_to_movie[movie_name].dropna().sort_values(ascending=False))
```

```python
corr_movie=corr_movie.join(ratings['number_of_ratings'])
```

```python
print("Predicted Movies")
corr_movie[corr_movie['number_of_ratings']>100].sort_values(movie_name,ascending=False).head(10)
```
